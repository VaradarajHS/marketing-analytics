```python
plt.hist(df["age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df["accountBalance"], bins=30)
plt.title("Account Balance")
plt.show()
```

```python
avg_balance = df.groupby("job")["accountBalance"].mean()
avg_balance.plot(kind="bar", figsize=(12,6))
plt.title("Average Balance by Job")
plt.show()
```
